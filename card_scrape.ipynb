```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import random
```

```python
# URL of the website
base_url = "https://pkmncards.com/page/{}/?s=&sort=date&ord=rev&display=list"

max_pages = 95
# Flag to track when to stop scraping
stop_scraping = False

# Lists to store scraped data
sets, numbers, names, rarities, prices = [], [], [], [], []
for page in range(1, max_pages + 1):
    # Format the URL with the current page number
    url = base_url.format(page)
    if stop_scraping:
        print("Reached 'LC' set. Stopping scrape.")
        break
    print(url)
    
    # Fetch the page content
    response = requests.get(url)
    if response.status_code != 200:
        print(f"Failed to retrieve page {page}. Skipping.")
        continue

    # Parse the HTML content
    soup = BeautifulSoup(response.content, "html.parser")


    # Locate the main container
    main_div = soup.find("main", class_='content')
    if not main_div:
        print("Could not find the main content area.")
        exit()

    # Extract card details
    cards = main_div.find_all("div", class_="entry-content")  # Replace with the actual class for card entries

    for card in cards:
        # Extract data for each card
        set_name = card.find("div", class_="column cell set")  # Adjust class name based on inspection
        card_number = card.find("div", class_="column cell number")
        card_name = card.find("div", class_="column cell name")
        rarity = card.find("div", class_="column cell rarity")
        price = card.find("div", class_="column cell price")

        # Check if the set name is "LC" and break if found
        #if set_name and set_name.get_text(strip=True) == "LC":
        #    stop_scraping=True
        #    break

        # Get text or default to "N/A"
        sets.append(set_name.get_text(strip=True) if set_name else "N/A")
        numbers.append(card_number.get_text(strip=True) if card_number else "N/A")
        names.append(card_name.get_text(strip=True) if card_name else "N/A")
        rarities.append(rarity.get_text(strip=True) if rarity else "N/A")
        prices.append(price.get_text(strip=True) if price else "N/A")



    # Wait for a random time between 1 to 3 seconds to avoid overloading the server
    time.sleep(random.uniform(1, 3))  # Random delay between 1 and 3 second

# Create DataFrame
data = pd.DataFrame({
    "Set": sets,
    "#": numbers,
    "Name": names,
    "Rarity": rarities,
    "Price": prices
})

# Save to CSV
data.to_csv("pokemon_cards.csv", index=False)
print("Scraping completed. Data saved to 'pokemon_cards.csv'.")
```
